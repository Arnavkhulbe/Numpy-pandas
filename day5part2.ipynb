```python
# Merging when Column Names are DIFFERENT

# or when there are multiple matching columns
import pandas as pd

dfA = pd.DataFrame({
    "Name": ["Amit", "Riya", "Karan", "Neha"],
    "Age": [16, 17, 16, 18]
})


dfB = pd.DataFrame({
    "Student": ["Amit", "Riya", "Karan", "Neha"],
    "Marks": [85, 78, 88, 95]
})
```

```python
merged = pd.merge(dfA, dfB, left_on="Name", right_on="Student")
print(merged) #Column duplication ho gaya: Name + Student
```

```python
merged.drop("Student", axis=1, inplace=True)
print(merged)
```

```python
merged = pd.merge(dfA, dfB, on="Name", suffixes=("_Left", "_Right")) 

#aur iska matlab ye hai ki dono table mai do same name column hai jinme se ek column same hai vo hai
# unique which is name and second is age which is not unique so usse left and right suffix dediya aur
# isme sirf voi rows aaegni jo dono table mai exist karti ho ek example ke sath btao
```
