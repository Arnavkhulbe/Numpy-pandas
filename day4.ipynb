```python
import pandas as pd

data = {
    "Name": ["Amit", "Riya", "Amit", "Karan", "Riya", "Amit", "Neha"],
    "Subject": ["Math", "Science", "English", "Math", "English", "Science", "Math"],
    "Marks": [85, 78, 90, 88, 74, 5, 95]
}

df = pd.DataFrame(data)
print("ðŸ”¹ Original Data:")
print(df)
```

```python
avg=df.groupby("Name")["Marks"].mean() #Find average marks of each student
print(avg)
```

```python
result=df.groupby(["Name","Subject"])["Marks"].mean() #Group by 2 columns it is multi level grouping 
print(result)
```

```python
result=df.groupby(["Name","Subject"])["Marks"].mean().reset_index() #Add .reset_index() for cleaner output
print(result)
```

```python
 summary=df.groupby("Name")["Marks"].agg(["min","max","mean","count"]) #Use .agg() for multiple summaries
print(summary)
```

```python
summary.sort_values(by="mean", ascending=False, inplace=True)
print(summary)
```

```python
summary=df.groupby("Name")["Marks"].agg(["mean","max"])
print(summary)
```
