```python
import pandas as pd

dfMarks = pd.DataFrame({
    "Student": ["Amit", "Riya", "Karan", "Neha"],
    "Math": [85, 78, 88, 95],
    "Science": [80, 75, 92, 90]
})
print(dfMarks)
```

```python
dfAttend = pd.DataFrame({
    "Student": ["Amit", "Riya", "Karan", "John"],
    "Attendance": [90, 85, 88, 80]
})
print(dfAttend)
```

```python
dfSports = pd.DataFrame({
    "Student": ["Amit", "Neha", "John", "Riya"],
    "Sports": [8, 10, 6, 7]
})
print(dfSports)
```

```python
merged=pd.merge(dfMarks,dfAttend,on="Student",how="outer")
print(merged)
```

```python
merge=pd.merge(merged,dfSports,on="Student",how="outer")
print(merge)
```

```python
merge.fillna(0, inplace=True)
print(merge)
```

```python
merge['Total'] = merge.drop(columns=['Student']).sum(axis=1) 
print(merge)
```

```python
topper=merge['Total'].idxmax()
print("top student: ",merge.loc[topper])
```
