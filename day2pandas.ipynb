```python
data={#assignment
"student":["A","B",] ,   
"math":[85,56],
"english":[92,74],
"science":[60,90]}
arr2=pd.DataFrame(data)
print(arr2)
```

```python
print("average marks of maths:",arr2["math"].mean())
print("average marks of science:",arr2["science"].mean())
print("average marks of english:",arr2["english"].mean())
print("\n")

print("highest marks in maths:",arr2["math"].max())
print("\n")

arr2["total"]=arr2[["math","english","science"]].sum(axis=1)
print(arr2)
print("\n")

print("student with highest total marks:",arr2["total"].idxmax())
print("\n")

topper_index = arr2["total"].idxmax()
print("Topper student:", arr2.loc[topper_index, ["student", "total"]])
```
