```python
df2 = pd.DataFrame({
    "Student": ["Amit", "Riya", "Karan", "Neha"],
    "Attendance": [90, 85, 88, 97]
})
print(df2)
```

```python
merged=pd.merge(df1,df2,on="Student")
print(merged)
```

```python
# üß† Mismatch kya hota hai?

# Row Student name ke basis par milti hai.

# "Amit" ‚úî ‚Üí dono tables me

# "Riya" ‚úî ‚Üí dono me

# "Karan" ‚úî ‚Üí dono me

# Neha ‚ùå ‚Üí sirf df1 (Marks) me

# John ‚ùå ‚Üí sirf df2 (Attendance) me

# üëâ Yahi mismatch kehte hain
# üëâ Aur joins ka kaam hai decide karna:
# kisko final table me rakhe aur kisko na rakhe
```

```python
df1 = pd.DataFrame({
    "Student": ["Amit", "Riya", "Karan", "Neha"],
    "Marks": [85, 78, 88, 95]
})
```

```python
df2 = pd.DataFrame({
    "Student": ["Amit", "Riya", "Karan", "John"],
    "Attendance": [90, 85, 88, 80]
})
```

```python
#first is inner join which say include only common 

inner = pd.merge(df1, df2, on="Student", how="inner")
print(inner)
```

```python
#second is left join-- Marks wali table (df1) priority

left=pd.merge(df1,df2,on="Student",how="left")
print(left)
```

```python
#third  is right  join-- attendance wali table (df1) priority
right = pd.merge(df1, df2, on="Student", how="right")
print(right)
```

```python
# Sab jaga se sab kuch
# ‚Üí NaN where mismatch
outer = pd.merge(df1, df2, on="Student", how="outer")
print(outer)
```

```python
#pehle data banate hain (indexes different rakh kar)
df_marks = pd.DataFrame({
    "Marks": [85, 78, 88, 95]
}, index=["Amit", "Riya", "Karan", "Neha"])

df_att = pd.DataFrame({
    "Attendance": [90, 85, 88, 97]
}, index=["Amit", "Riya", "Karan", "Neha"])

print(df_marks)
print(df_att)
```

```python
joined = df_att.join(df_marks) #joined using attendance
print(joined)
```

```python
#concat() and row wise concat 
row_added = pd.concat([df_marks, df_marks])
print(row_added)
```

```python
col_added = pd.concat([df_marks, df_att], axis=1)
print(col_added) #columns wise concat 
```

```python
joined.to_csv("joined.csv", index=False)
```
