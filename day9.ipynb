```python
import pandas as pd

customers = pd.DataFrame({
    "CustomerID": ["C01", "C02", "C03", "C04"],
    "Name": ["Amit", "Riya", "Karan", "Neha"]
})

orders = pd.DataFrame({
    "OrderID": ["O01", "O02", "O03", "O04", "O05"],
    "CustomerID": ["C01", "C02", "C01", "C04", "C05"],
    "Amount": [500, 700, 300, 900, 600]
})
```

```python
orders
```

```python
customers
```

```python
merged_inner = pd.merge(customers, orders, on="CustomerID", how="inner") #inner join lagaya 
print(merged_inner)
```

```python
merged_left = pd.merge(customers, orders, on="CustomerID", how="left") #left join to check order missing
print(merged_left)
```

```python
merged_right = pd.merge(customers, orders, on="CustomerID", how="right") #we checked that customer missing 
print(merged_right)
```

```python
merged_outer = pd.merge(customers, orders, on="CustomerID", how="outer")
print(merged_outer) #har cheez dikhao jaha order ya cutomer missing everything
```

```python
print("number of total customer is: ",merged_outer["CustomerID"].nunique())
```

```python
merged_outer['OrderID'].isna() #to check which row has NaN value in orderid column
```

```python

print("Customers with Missing Orders:")
merged_outer[merged_outer["OrderID"].isnull()][["CustomerID","Name"]]

```

```python
suspicious = merged_outer[merged_outer["Name"].isnull()][["OrderID","Amount"]]
print("Suspicious Orders:")
print(suspicious)
```

```python
top_customer = merged_outer.groupby("Name")["Amount"].sum().sort_values(ascending=False)
print("Top Customer Spendings:")
print(top_customer)
```
