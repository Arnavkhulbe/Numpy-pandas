```python
import pandas as pd #Duration Calculate Karna

data = {
    "Start_Date": ["2025-01-01", "2025-02-01", "2025-03-10"],
    "End_Date": ["2025-01-10", "2025-02-20", "2025-04-05"]
}

df = pd.DataFrame(data)

df["Start_Date"] = pd.to_datetime(df["Start_Date"])
df["End_Date"] = pd.to_datetime(df["End_Date"])

df["Duration_days"] = (df["End_Date"] - df["Start_Date"]).dt.days

print(df)
```

```python
import pandas as pd
from datetime import datetime

data = {"Name": ["Amit", "Riya", "Neha"],
        "Birthdate": ["2000-05-20", "1998-10-15", "2003-01-30"]}

df = pd.DataFrame(data)

df["Birthdate"] = pd.to_datetime(df["Birthdate"])
today = pd.to_datetime("today")

df["Age"] = (today - df["Birthdate"]).dt.days // 365

print(df) #Calculate Age or Tenure from Date
```

```python
import pandas as pd #Date Based Filtering & Grouping
#goal- Kaunse din / kaunse month me sales zyada hoti hai?

data = {
    "Date": ["2025-01-01", "2025-01-03", "2025-01-05", "2025-01-07", "2025-01-10"],
    "Sales": [100, 200, 150, 300, 250]
}

df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])
df["Weekday"] = df["Date"].dt.day_name()
df["Month"] = df["Date"].dt.month
print(df)
```

```python
weekend = df[df["Weekday"].isin(["Saturday", "Sunday"])]
print(weekend) #Matlab â†’ Weekend me sirf 1 sale hui (150)
```

```python
result = df.groupby("Weekday")["Sales"].mean()
print(result)   # neeche output ke hisaab se tuesday ko sabse best sale hua 
```

```python
monthly_sales = df.groupby("Month")["Sales"].sum()
print(monthly_sales) #har month ki total sales add karta hai
```
