```python
import pandas as pd

data = {
    "Name": ["Amit ", "Riya", "ARJUN", "neha", "Karan", "Unknown"],
    "Math_Score": [85, 78, 90, 67, "88", None],
    "Science_Score": [88, "81", 95, "N/A", 77, 69],
    "English_Score": [92, 74, "89", 70, 91, "missing"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.rename(columns={ #changing column name 
    "Math_Score": "maths",
    "Science_Score": "science",
    "English_Score": "english"},inplace=True)
print(df.columns)
```

```python
import numpy as np #Replace wrong string values with NaN

df.replace(["N/A", "missing", None], np.nan, inplace=True)
print(df)
```

```python
df["maths"]=pd.to_numeric(df["maths"],errors='coerce')     #Convert column types (numbers as strings â†’ float)
df["science"]=pd.to_numeric(df["science"],errors='coerce')
df["english"]=pd.to_numeric(df["english"],errors='coerce')
```

```python
df.info()
```

```python
df["Name"]=df["Name"].str.strip()   #Remove extra spaces
df["Name"] = df["Name"].str.title()  #Convert all names to title case
```

```python
df
```

```python
df["maths"].fillna(df["maths"].mean(),inplace=True)#Fill missing marks with mean
```

```python
df = df.round(1)
print(df)
```

```python
import pandas as pd

data = {
    "Name": ["Amit", "Riya", "Arjun", "Riya", "Karan", "Amit"],
    "Math": [85, 78, 90, 78, 88, 85],
    "Science": [88, 81, 95, 81, 77, 88],
    "English": [92, 74, 89, 74, 91, 92]
}

df = pd.DataFrame(data)
print("ðŸ”¹ Original Data:")
print(df)
```

```python
df.duplicated() #checks duplicate rows 
```

```python
df.duplicated().sum()#counts duplicate rows 
```

```python
df[df.duplicated()] #shows only duplicate rows 
```

```python
df1=df.drop_duplicates() #removes duplicate rows 
print(df1)
```

```python
df2=df.drop_duplicates(keep='last') #deletes all duplicate row but keeps the last one
df2
```

```python
df2.reset_index(drop=True,inplace=True) #fixed indexing
print(df2)
```
