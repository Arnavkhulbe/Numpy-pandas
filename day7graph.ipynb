```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
students = ["Amit", "Riya", "Karan", "Neha"]
marks = [85, 78, 88, 95]
```

```python
plt.plot(students, marks)      #Basic Plotting (Line Graph)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

```

```python
import matplotlib.pyplot as plt #Part 2 â€” Bar Chart (Comparison Graph)

students = ["Amit", "Riya", "Karan", "Neha"]
marks = [85, 78, 88, 95]
```

```python
plt.bar(students, marks)
plt.title("Student Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
```

```python
import matplotlib.pyplot as plt
students = ["Amit", "Riya", "Karan", "Neha"]
marks = [85, 78, 88, 95]
plt.pie(marks, labels=students, autopct="%1.1f%%")
plt.title("Marks Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

study_hours = [2, 3, 4, 5, 6]
marks = [50, 60, 65, 75, 90]   #Scatter Plot
# Ye graph Machine Learning me sabse important visualization hota hai
# kyunki ye relationship dikhata hai between two variables 
```

```python
plt.scatter(study_hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.show()
```

```python
# Histogram data ke distribution ka graph hai.
# Mtlb â†’ Data zyada kis range me hai?
# Kaun si value zyada baar repeat ho rahi hai?


import matplotlib.pyplot as plt

marks = [85, 78, 88, 95, 65, 72, 91, 84, 77, 69, 55, 60]

plt.hist(marks, bins=5)
plt.title("Marks Distribution")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Outliers identify kare
# Kya abnormal values hain jisse model disturb ho?

import matplotlib.pyplot as plt

marks = [55, 60, 65, 69, 72, 77, 78, 84, 85, 88, 91, 10]

plt.boxplot(marks)
plt.title("Marks Spread & Outliers Check")
plt.ylabel("Marks")
plt.show()
```
